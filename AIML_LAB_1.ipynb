```python
import matplotlib.pyplot as plt
x=[1,2,3]
y=[10,20,30]
z=[10,40,60]
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.plot(x,y,z,label='3d')
ax.set_xlabel("year")
ax.set_ylabel("Rate")
ax.set_zlabel("number")

plt.show()
```
